```python
from keras.models import load_model

model = load_model('cvd.h5')
```

```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
img=load_img('test/65.jpg',target_size=(150,150))
```

```python
img
```

```python
img=img_to_array(img)
```

```python
img.shape
```

```python
img1=img.reshape(1,150,150,3)
```

```python
img1.shape
```

```python
model.predict(img1)
```

<h1>FLow</h1><b>
<ol>Steps:
    <li>Prepare the dataset, in this case the images. We make three folders: train, validation &amp; test. The train &amp; validation folders further contain folder names with the labels, i.e, cat &amp; dog. The test folder contains images to be predicted. The validation folder contain images where the weights obtained from training images are applied on to get accuracy metrics.</li>
    <li>The images are reshaped to size(150,150,3).</li>
    <li>The model is created.</li>
    <li>We train the model on the images. This is achieved using the keras imagegenerator class which clubs the training &amp; validation images to a 2D array. Image generator is also applied to testing folder.</li>
    <li>We save our weights in a .h5 file. So that we don't have to train our model everytime. To predict, we just load the model and fashion the test image in the form (1,150,150,3).</li>
    <li>We record the predicted result. The result is in the form of probability matrix of size (1,2). The first data is probability of "cat" &amp; the second is the probability of "dog".</li>
    </ol>
    </b>
